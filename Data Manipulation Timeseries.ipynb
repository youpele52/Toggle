```python
import pandas as pd
from datetime import datetime
import matplotlib.pyplot as plt
from matplotlib import pyplot
import math
```

# Importing the data


```python
series = pd.read_csv('./Assets/eurcad_price.csv', header= None,  index_col=0, parse_dates=True, squeeze=True)
```

Visualising the data

```python
series.head
```

```python
series.plot.line()
pyplot.show()
```

# Calculating Annualized Standard Deviation for 250 trading days in the year.

```python
std = series.std(skipna=True)
annual_standard_deviation = std * math.sqrt(12)

print('Standard Deviation:', std,'\nAnnualized Standard Deviation:' ,annual_standard_deviation)
```

 # Calcluating 1M annualized standard deviation.

```python
series_M=series.resample('1M').mean()
series_M.head
```

```python
series_M.plot()
```

```python
std_M = series_M.std(skipna=True)
annual_standard_deviation = std_M * math.sqrt(12)

print('Standard Deviation:', std_M,'\n1M Annualized Standard Deviation:' ,annual_standard_deviation)
```
